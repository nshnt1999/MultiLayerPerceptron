```python
def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p))
    return l


def one_hot(y,depth):
    m = y.shape[0]
    oht = np.zeros((m,depth))
    oht[np.arange(m),y]=1
    return oht
```

## Generating data

```python
#make_circles(n_samples=100, shuffle=True, noise=None, random_state=None, factor=0.8)
X,y = make_circles(n_samples=500,shuffle=True,noise=0.2,random_state=1,factor=0.3)
```

## Training of Model

```python
model = NeuralNetwork(2,[10,5],2)
```

```python
def train(X,Y,model,epochs,learning_rate,logs=True):
    training_loss = []
    
    classes = 2
    Y_OHT = one_hot(Y,classes)
    
    for ix in range(epochs):
        Y_ = model.forward_propogation(X)
        l = loss(Y_OHT,Y_)
        training_loss.append(l)
        model.backward_propogation(X,Y_OHT,learning_rate)
        
        if(logs):
            print("Epoch %d Loss %.4f"%(ix,l))
            
    
    return training_loss
```

```python
l1 = train(X,y,model,epochs=500,learning_rate=0.001,logs=True)
```
